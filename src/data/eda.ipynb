```python

def checkSeason(month):
  if month in [12, 1, 2]:
    return 'Winter'
  elif month in [3, 4, 5]:
    return 'Spring'
  elif month in [6, 7, 8]:
    return 'Summer'
  else:
    return 'Fall'

df['SeasonSold'] = df['MoSold'].apply(checkSeason)
df['HouseAge'] = df['YrSold'] - df['YearBuilt']

df['AmenitiesPresence'] = df.apply(
    lambda row: 1 if (
        (row['GarageQual'] != 'NA') or
        (row['Fireplaces'] > 0) or
        (row['MiscVal'] > 0)
    ) else 0,
    axis=1
)

df['AmenitiesQuantity'] = df.apply(
    lambda row:
        int(row['GarageQual'] != 'NA') +
        int(row['Fireplaces'] > 0) +
        int(row['MiscVal'] > 0),
    axis=1
)

numerical_features.append('AmenitiesQuantity')
numerical_features.append('HouseAge')
df_visualization = df.copy()   # Kopi disini sebelum di scaling/encoding untuk section kedua
```

```python
def preprocessing_pipeline(df): 
    """
    For inference we create the pipeline to do all the things we did above (we assume the df has the same columns as the one above)
    """
    numerical_features = df.select_dtypes(include=['number']).columns.tolist()  
    
    df = df[df['LotFrontage'] <= 300]
    df = df[df['LotArea'] <= 100000]
    df = df[df['BsmtFinSF1'] <= 5000]
    df = df[df['1stFlrSF'] <= 4000]
    df = df[df['TotalBsmtSF'] <= 6000]
    df = df[df['SalePrice'] <= 700000]
    df.drop(columns=['PoolQC', 'MiscFeature', 'Alley', 'Fence'], inplace=True)
    
    for col in [x for x in df.columns.tolist() if x not in numerical_features]:
        df[col] = df[col].fillna('NA')
    
    imputer = KNNImputer()
    imputer = KNNImputer(n_neighbors=4)
    y = imputer.fit_transform(df.loc[:, ['MasVnrArea']])
    df['MasVnrArea'] = y.round(0)

    df['GarageYrBlt'] = df['GarageYrBlt'].fillna(df['GarageYrBlt'].median())

    
    def checkSeason(month):
        if month in [12, 1, 2]:
            return 'Winter'
        elif month in [3, 4, 5]:
            return 'Spring'
        elif month in [6, 7, 8]:
            return 'Summer'
        else:
            return 'Fall'

    df['SeasonSold'] = df['MoSold'].apply(checkSeason)
    df['HouseAge'] = df['YrSold'] - df['YearBuilt']

    df['AmenitiesPresence'] = df.apply(
        lambda row: 1 if (
            (row['GarageQual'] != 'NA') or
            (row['Fireplaces'] > 0) or
            (row['MiscVal'] > 0)
        ) else 0,
        axis=1
    )

    df['AmenitiesQuantity'] = df.apply(
        lambda row:
            int(row['GarageQual'] != 'NA') +
            int(row['Fireplaces'] > 0) +
            int(row['MiscVal'] > 0),
        axis=1
    )

    numerical_features.append('AmenitiesQuantity')
    numerical_features.append('HouseAge')


    # Ordinal encoding
    ordinal = ['LotShape', 'Utilities', 'LandSlope',
            'ExterQual', 'ExterCond', 'BsmtQual', 'BsmtCond','BsmtExposure',
            'BsmtFinType1', 'BsmtFinType2', 'HeatingQC', 'KitchenQual',
            'Functional', 'FireplaceQu', 'GarageFinish',
            'GarageQual', 'GarageCond']

    ordinal_categories = [
        ['IR3', 'IR2', 'IR1', 'Reg'],  # LotShape
        ['ELO', 'NoSeWa', 'NoSewr', 'AllPub'],  # Utilities
        ['Sev', 'Mod', 'Gtl'],  # LandSlope
        ['Po', 'Fa', 'TA', 'Gd', 'Ex'],  # ExterQual
        ['Po', 'Fa', 'TA', 'Gd', 'Ex'],  # ExterCond
        ['NA', 'Po', 'Fa', 'TA', 'Gd', 'Ex'],  # BsmtQual
        ['NA', 'Po', 'Fa', 'TA', 'Gd', 'Ex'],  # BsmtCond
        ['NA', 'No', 'Mn', 'Av', 'Gd'],  # BsmtExposure
        ['NA', 'Unf', 'LwQ', 'Rec', 'BLQ', 'ALQ', 'GLQ'],  # BsmtFinType1
        ['NA', 'Unf', 'LwQ', 'Rec', 'BLQ', 'ALQ', 'GLQ'],  # BsmtFinType2
        ['Po', 'Fa', 'TA', 'Gd', 'Ex'],  # HeatingQC
        ['Po', 'Fa', 'TA', 'Gd', 'Ex'],  # KitchenQual
        ['Sal', 'Sev', 'Maj2', 'Maj1', 'Mod', 'Min2', 'Min1', 'Typ'],  # Functional
        ['NA', 'Po', 'Fa', 'TA', 'Gd', 'Ex'],  # FireplaceQu
        ['NA', 'Unf', 'RFn', 'Fin'],  # GarageFinish
        ['NA', 'Po', 'Fa', 'TA', 'Gd', 'Ex'],  # GarageQual
        ['NA', 'Po', 'Fa', 'TA', 'Gd', 'Ex'],  # GarageCond
    ]
    encoder = OrdinalEncoder(categories=ordinal_categories, handle_unknown='use_encoded_value', unknown_value=-1)
    df[ordinal] = encoder.fit_transform(df[ordinal])
    df[ordinal]
    
    minmax_columns = ['HouseAge', 'YrSold', 'YearBuilt', 'GarageYrBlt', 'YearRemodAdd', 'MoSold', 'AmenitiesQuantity']
    standard_columns = [col for col in scaling_columns if col not in minmax_columns]

    minmax_scaler = MinMaxScaler()
    standard_scaler = StandardScaler()

    df[standard_columns] = standard_scaler.fit_transform(df[standard_columns])
    df[minmax_columns] = minmax_scaler.fit_transform(df[minmax_columns])

    cols_to_keep = ['OverallQual', 'YearBuilt', 'YearRemodAdd', 'ExterQual', 'BsmtQual', 'TotalBsmtSF', '1stFlrSF', 'GrLivArea', 'FullBath', 'KitchenQual', 'TotRmsAbvGrd', 'FireplaceQu', 'GarageFinish', 'GarageCars', 'GarageArea', 'HouseAge']

    return df[cols_to_keep]

    


    

```
